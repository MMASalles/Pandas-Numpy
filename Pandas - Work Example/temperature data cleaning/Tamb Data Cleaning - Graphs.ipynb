# Check irradiance data

```python
def Irradiance_Graph(data, title):

    # Plot graph
    fig, ax = plt.subplots(figsize=(35,5))
    ax.plot(data)

    ax.set_title('SMR - Average MET Tamb - Cleaned Data - ' + title, size=22)

    # Format the ticks
    ax.xaxis.set_major_locator(mdates.DayLocator()) # every month
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%d')) # show date on major grid marker
    ax.xaxis.set_minor_locator(mdates.HourLocator()) # every day
    #ax.xaxis.set_minor_formatter(mdates.DateFormatter('%H')) # show hour on minor grid marker

    # Round to nearest hour.
    datemin = np.datetime64(data.index[0], 'h')
    datemax = np.datetime64(data.index[-1], 'h')
    ax.set_xlim(datemin, datemax)
    ax.set_ylim((-10,50))

    # Grid, labels & legend
    plt.grid(True)
    plt.xlabel('Date & Time', fontsize=18)
    plt.ylabel('Temperature (Â°C)', fontsize=18)

    #plt.savefig(title+'.png')
    plt.show()
    
```

### DEC

```python
month_filter = (Data.index >= '2019-12-01') & (Data.index < '2020-01-01')
title = 'DEC-19'
Irradiance_Graph(Data.loc[month_filter, 'Step 5'], title)
```

### JANUARY

```python
month_filter = (Data.index >= '2020-01-01') & (Data.index < '2020-02-01')
title = 'JAN-20'
Irradiance_Graph(Data.loc[month_filter, 'Step 5'], title)
```
